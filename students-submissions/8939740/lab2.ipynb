# PART-A

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB, MultinomialNB
import numpy as np
```

```python
lab2_df = pd.read_csv('Lab2_dataset.csv')
lab2_df.head(10)
vectorizer = CountVectorizer()
text_vector = vectorizer.fit_transform(lab2_df['text'])
text_vector.toarray()
```

```python
X_train, X_test, y_train, y_test = train_test_split(text_vector, lab2_df['label'], test_size=0.2, random_state=30)
svc_model = SVC()
svc_model.fit(X_train, y_train)
svc_predictions = svc_model.predict(X_test)
svc_accuracy = accuracy_score(y_test, svc_predictions)
print("SVC Model Accuracy :", svc_accuracy)
```

```python
gaussian_nb = GaussianNB()
gaussian_nb.fit(X_train.toarray(), y_train)
gaussian_nb_predictions = gaussian_nb.predict(X_test.toarray())
gaussian_nb_accuracy = accuracy_score(y_test, gaussian_nb_predictions)
print("Gaussian Naive Bayes Accuracy :", gaussian_nb_accuracy)
multinomial_nb = MultinomialNB()
multinomial_nb.fit(X_train, y_train)
multinomial_nb_predictions = multinomial_nb.predict(X_test)
multinomial_nb_accuracy = accuracy_score(y_test, multinomial_nb_predictions)
print("Multinomial Naive Bayes Accuracy :", multinomial_nb_accuracy)
```

```python
lab2nyc_df = pd.read_csv('AB_NYC_2019.csv')
lab2nyc_df.head(10)
```

```python


price_per_night = lab2nyc_df['price']

z_score = np.abs((price_per_night - price_per_night.mean()) / price_per_night.std())
threshold = 7
cleaned_zscore = lab2nyc_df[z_score <= threshold]

Q1 = price_per_night.quantile(0.25)
Q3 = price_per_night.quantile(0.75)
IQR = Q3 - Q1
lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR
cleaned_whiskers = lab2nyc_df[(price_per_night >= lower_whisker) & (price_per_night <= upper_whisker)]

print("Number of rows in the original dataset:", len(lab2nyc_df))
print("Number of rows after removing outliers using Z-score approach:", len(cleaned_zscore))
print("Number of rows after removing outliers using whiskers approach:", len(cleaned_whiskers))
```

## Here whiskers approach is better because it removes more outliers
